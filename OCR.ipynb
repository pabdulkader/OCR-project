```python
from io import IncrementalNewlineDecoder
#step 1 import necessary libraries
import pytesseract
from PIL import Image
from google.colab import files
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
#step 2 define a fuction to perform OCR
def perform_ocr(image_path):
  try:
    img=Image.open(image_path)
    text=pytesseract.image_to_string(img)
    return text
  except Exception as e:
    return f"error during OCR:{e}"
```

```python
#step 4 upload the image containing text
uploaded=files.upload()

if uploaded:
  uploaded_file_name=list(uploaded.keys())[0]
  print(f"uploaded image: (uploaded_file_name)")

  #optional display the uploaded image
  img=cv2.imread(uploaded_file_name)
  plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
  plt.title("uploaded image")
  plt.axis("off")
  plt.show()

  extracted_text=perform_ocr(uploaded_file_name)
  print(extracted_text)
else:
  print("no image uploaded")
```
